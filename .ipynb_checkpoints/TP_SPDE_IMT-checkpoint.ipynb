# I. Introduction

```r
# Cleaning the R environment
rm(list = ls())
options(warn=-1)

# Loading libraries
library(Matrix)
library(fields)
library(viridis)
library(gifski)
library(IRdisplay)
library(pracma)
library(raster)

## Square difference between two vectors
diffSq<-function(x,y){
  (x-y)^2
}

## define sample variogram
expVar<-function(Z,n0,lag,nbpts,Nx,Ny,dx,dy){
  
  kech=sample(1:(Nx*Ny),n0)
  xech=((kech-1)%%Nx+1)*dx
  yech=((kech-1)%/%Nx+1)*dy
  nmax=min(length(xech),length(yech))
  xech=xech[1:nmax]
  yech=yech[1:nmax]
    
  dist=(sqrt(outer(xech,xech,diffSq)+outer(yech,yech,diffSq)))
  dist=dist[upper.tri(dist)]
  diffZ=(outer(Z[kech],Z[kech],diffSq))
  diffZ=diffZ[upper.tri(diffZ)]
  
  breaks=seq(from=0,by=lag,length.out = nbpts)
  lagcut=cut(dist, breaks=breaks, labels =FALSE)
  varexp=tapply(diffZ,lagcut,mean)/2
  return(cbind(lag=lag/2+(as.numeric(rownames(varexp))-1)*lag,varexp))
}

## Matérn covariance
Matern<-function(x,kappa,nu,cov=FALSE){
  if(cov) return((2^(1-nu)/gamma(nu))*(sqrt(2*nu)*kappa*x)^(nu)*besselK(sqrt(2*nu)*kappa*x,nu))
  else    return(1-(2^(1-nu)/gamma(nu))*(sqrt(2*nu)*kappa*x)^(nu)*besselK(sqrt(2*nu)*kappa*x,nu)) 
}
```

While most of Geostatistical estimation procedures suffer from severe limitations when considering large dimensions for both observation vector $\mathbb{R}^p$ and state space $\mathbb{R}^m$, the SPDE-based estimation model is a new formalism introduced by \citep{LINDGREN_2011} constrained by the use of Matèrn covariance model:
\begin{align*}
 C(\mathbf{h})=\sigma^2\frac{2^{1-\nu}}{\Gamma(\nu)}\left(\sqrt{2\nu}\frac{h}{a}\right)^\nu \mathcal{K}_\nu\left(\sqrt{2\nu}\frac{h}{a}\right)
\end{align*}

with:

* $h=|\mathbf{s}-\mathbf{s}'|$ is the distance between two spatial locations
* $\mathcal{K}_\nu$ is the modified 2nd order Bessel function
* $\nu$ controls the regularity of the covariance

Let note that for:

* $\nu=1/2$, we get the exponential covariance
* $\nu\rightarrow +\infty$, we get the gaussian covariance

If $\mathbf{X}(\mathbf{s})$, $\mathbf{s} \in \mathbb{R}^d $, is a Gaussian random field (GRF) with Matern covariance having regularity parameters \nu, scale parameter $a$ and  variance $\sigma^2$, then $\mathbf{X}(\mathbf{s})$ is solution of the following SPDE \citep{WHITTLE_1953}:
\begin{align}
\label{isotropic_SPDE}
(\kappa^2 - \Delta)^{\alpha/2} \mathbf{X}(\mathbf{s}) = \tau\mathbf{Z}(\mathbf{s})
\end{align} 

with:

* $\Delta = \sum_{i=1}^d \frac{\partial^2}{\partial s_i^2}$ the Laplacian
* $\mathbf{Z}(\mathbf{s})$ is standard Gaussian white noise
* $\kappa = 1/a$
* $\alpha = \nu + d/2$
* $\tau = \frac{\sigma\Gamma(\nu+d/2)^{1/2}(4\pi)^{d/4}\kappa^{\nu}}{\Gamma(\nu)^{1/2}}$

```r
plot(Matern(seq(0,50,.1),.2,10,cov=TRUE),col="red",type='l',ylim=c(0,1),ylab='C(h)',xlab='h')
lines(Matern(seq(0,50,.1),.2,1/2,cov=TRUE),col="blue",type='l')
legend('topright', legend=c("Gaussian", "Exponential"),col=c("red", "blue"), lty=1:2, cex=1.5,box.col=0)
plot(Matern(seq(0,50,.1),.2,10,cov=FALSE),col="red",type='l',ylim=c(0,1),ylab='gamma(h)',xlab='h')
lines(Matern(seq(0,50,.1),.2,1/2,cov=FALSE),col="blue",type='l')
legend('bottomright', legend=c("Gaussian", "Exponential"),col=c("red", "blue"), lty=1:2, cex=1.5,box.col=0)
```

## II. Spatial case

Let now consider state space $\mathbf{x} \in \mathbb{R}^m$, which can be seen as a discretized version of process $\mathbf{x}(\mathbf{s})$ on a regular grid of size $m = N_x \times N_y$ with grid step $dx$ and $dy$. Then, an element of state space $\mathbf{x}$ can be located on the grid either by its coordinates $(x, y)$ or by indices $(i, j)$ where $i \in \lbrace{1, \cdots, N_x \rbrace}$ and $j \in \lbrace{1,\cdots,N_y \rbrace}$.

A discretized version of SPDE \ref{isotropic_SPDE} on a  grid node $(i, j)$ is:
\begin{eqnarray*}
\kappa^2-\Delta\mathbf{x}(i,j)=\tau \mathbf{z}(i,j)
\end{eqnarray*} 
is obtained using a centered finite difference scheme of the Laplacian operator $\Delta=\frac{\partial{}}{\partial{x}} + \frac{\partial{}}{\partial{y}}$. We will approach the partial derivative of $\mathbf{x}$ with respect to coordinate $y$ by a centered difference: 
\begin{eqnarray*}
\frac{\partial \mathbf{x}}{\partial y}=\frac{\mathbf{x}(y+dy/2)-\mathbf{x}(y-dy/2)}{dx}
\end{eqnarray*}

We get: 
\begin{eqnarray*}
\kappa^2 \mathbf{x}(i,j) &-&\frac{\mathbf{x}(i-1,j)-2\mathbf{x}(i,j)+\mathbf{x}(i+1,j)}{dx^2} \\
&-& \frac{\mathbf{x}(i,j-1)-2\mathbf{x}(i,j)+\mathbf{x}(i,j+1)}{dy^2} = \mathbf{z}(i,j)
\end{eqnarray*}

## II.1 DF Discretized Laplacian

Let now write the discretized of version SPDE \ref{isotropic_SPDE} in matrix notations by using the vectors $\mathbf{x}$ and
$\mathbf{z}$ and a matrix $\mathbf{A}$ corresponding to the discretized version of fractional differential operator over the grid:
\begin{align*}
\mathbf{A}\mathbf{x}=\tau\mathbf{z}
\end{align*} 
or :
* $A_{i,j}$ = {$-1/dx^2$} if $j=i+1$ or $j=i-1$
* $A_{i,j}$ = {$-1/dy^2$} if $j=i+N_x$ or $j=i-N_x$
* $A_{i,j}$ = $\kappa^2 + 2(1/dx^2+1/dy^2)$ if $j=i$
* $A_{i,j}$ = 0 otherwise

```r
# Discretized Laplacian (differential operator)
## Inputs :  
## Nx : number of points in direction x
## Ny : number of points in direction y
## dx : lag in direction x
## dy : lag in direction y
## Results :
## Sparse matrix with size (Nx*Ny)x(Nx*Ny) 

DiffOperator<-function(Nx,Ny,dx,dy, kappa){
  
  nbNodes=Nx*Ny
  indices=1:nbNodes
  

  # We start by creating an array containing the indices of the non-zero elements
  # of A with associated values and we will use the R function sparseMatrix 
  # to create A from this array

  # Right neighbour ( !! remove last column)
  indicesVoisins=indices[-which(indices%%Nx==0)]
  # Matrix values
  k1=cbind(indicesVoisins,indicesVoisins+1, -1/dx^2)
  
  # Left neighbour ( !! remove first column)
  indicesVoisins=indices[-which(indices%%Nx==1)]
  # Matrix values
  k2=cbind(indicesVoisins,indicesVoisins-1, -1/dx^2)
  
  # Top neighbour  ( !! remove last line )
  indicesVoisins=indices[-which(indices>(Ny-1)*Nx)]
  # Matrix values
  k3=cbind(indicesVoisins,indicesVoisins+Nx, -1/dy^2)
  
  # Bottom neighbour ( !! remove first line )
  indicesVoisins=indices[-which(indices<Nx+1)]
  # Matrix values
  k4=cbind(indicesVoisins,indicesVoisins-Nx, -1/dy^2)
  
  # Central point
  k5=cbind(indices,indices,kappa^2+2*(1/dx^2+1/dy^2))

  # Stack all neighbours
  k=rbind(k1,k2,k3,k4,k5)
  
  # Create sparseMatrix
  return(sparseMatrix(i=k[,1],j=k[,2],x=k[,3],dims = c(nbNodes,nbNodes)))
  
}

# Test function
A=DiffOperator(10,10,1,1,0)
image(A)

## Degree of differential operator
pow_diff_operator<-function(A,pow){
    B=A
    if(pow>1){
      for(i in 1:(pow-1)){
        B=B%*%A   
      }
    }
    # remove border symmetry issues
    B=(1/2)*(t(B)+B)
    return(B)
}
```

## II.2 Run a simulation

We can easily show that the discretization matrix $\mathbf{A}$ of
the differential operator is invertible. Regarding the precision matrix
$\mathbf{Q}$ of $\mathbf{x}$ (i.e. the inverse of its covariance matrix of
covariance), it is:
\begin{align*}
&\mathbf{A}\mathbf{x}=\tau\mathbf{z} \\
&\mathrm{Cov}\left\lbrace \mathbf{A}\mathbf{x},\mathbf{A}\mathbf{x} \right\rbrace= \mathbf{A} \mathrm{Cov}\left\lbrace \mathbf{x},\mathbf{x} \right\rbrace \mathbf{A}^\mathrm{T} = \mathrm{Cov}\left\lbrace \tau\mathbf{z},\tau\mathbf{z} \right\rbrace \\
& \mathbf{A} \mathbf{\Sigma} \mathbf{A}^\mathrm{T} = \frac{\tau^2}{dxdy}\mathbf{I} \\
& \mathbf{\Sigma} = \frac{\tau^2}{dxdy} (\mathbf{A})^{-1} (\mathbf{A}^\mathrm{T})^{-1} \\
& \mathbf{Q} = \mathbf{\Sigma}^{-1} = \frac{dxdy}{\tau^2}\mathbf{A}^\mathrm{T}\mathbf{A}
\end{align*}

A non-conditional simulation of state $\mathbf{x}$ is obtained by solving the linear system introduced above for a random vector $\mathbf{z}$ whose components are independent Gaussians of variance $\tau^2/dxdy$.

```r
# Raster-based image plot
spplot<-function(U,zlim,title,display_vario=FALSE,cmap=viridis(128)){
  .pardefault <- par()
  if(display_vario) layout(matrix(c(1,2),ncol=2), widths=c(2,2), heights=c(1,1), FALSE)
  U[U<=zlim[1]]=zlim[1]
  U[U>=zlim[2]]=zlim[2]
  ras=raster(list(x=1:Nx,y=1:Ny,z=matrix(U,Nx)),crs=NA)
  plot(ras,col=cmap,zlim=zlim)
  title(title)
  if(display_vario) {
    lag=1
    varexp=expVar(U,n0=Nx*Ny/2,lag = lag, nbpts = floor(0.6*Nx/lag),Nx,Ny,dx,dy)
    plot(varexp[,1],varexp[,2],type="l",ylim=c(0,1.2),xlab='h',ylab='gamma(h)')
    points(varexp[,1],varexp[,2])
    curve(Matern(x,kappa,nu),add=TRUE,col="red")   
    par(.pardefault)
  }
}

# parameters
## grid parameters
resolution=5
Nx=Ny=10*resolution;
nbNodes=Nx*Ny
dx=dy=1;
## SPDE parameters
kappa=.33            # (1/a) scale parameter
pow=2                # degree of differential operator 
d=2                  # space dimension
alpha  = pow*2       # 
nu     = alpha-d/2   # regularity
sig2  = gamma(nu)/(gamma(nu+d/2)*((4*pi)^(d/2))*(kappa^(2*nu))) # variance of the matern field
tau=( (gamma(nu+d/2)*((4*pi)^(d/2))*(kappa^(2*nu))) / gamma(nu) )^(1/2) # 1/sd Matern fields
#print(sig2)

# Discrete differential operator
A=DiffOperator(Nx,Ny,dx,dy,kappa)
B=pow_diff_operator(A,pow)

# Function to generate a non-conditional GRF simulation
## Inputs : 
## A: Discrete differential operator
## dx,dy : pas de la grille de travail
## Results
## Gaussian solution of discretized SPDE
simuNC<-function(A,dx,dy,tau){ 
  nbNodes=dim(A)[1]
  eps=(tau/sqrt(dx*dy))*rnorm(nbNodes,0,1)
  zc=solve(A,eps)
  return(zc) 
}

# Non conditional simulation
Z = simuNC(B,dx,dy,tau)
# Display
spplot(matrix(Z,nrow = Nx,ncol = Ny),zlim=c(-2,2),title="Non conditional simulation",display_vario=FALSE)

check=FALSE
if(check){
  ## Matérn field comparison
  # Mean of sample variogram over 100 simulations
  nbSim=100          # number of simulations
  n0=0.2*(Nx*Ny)     # number of points used for the variogram
  lag=1              # Lag
  nbpts=0.75*Nx/lag  # nLag
  varexp=0
  for(i in 1:nbSim){
    Z0 = simuNC(B,dx,dy,tau)
    varexp=varexp+expVar(Z0,n0=n0,lag=lag,nbpts =nbpts,Nx,Ny,dx,dy)/nbSim
  }
  rm(Z0)
  # Display variogram
  plot(varexp[,1],varexp[,2],type="o", ylim=c(0, 1.2),pch=19)
  curve(Matern(x,kappa,nu,cov=FALSE),add=TRUE,col="red")
}
```

## II.3 Kriging

Regarding optimal interpolation, let note $\mathbf{y}$ the observations, the covariance matrix of \begin{bmatrix} \mathbf{y} \\ \mathbf{x} \end{bmatrix} is:
\begin{align*}
\mathbf{\Sigma} = \begin{bmatrix} \mathbf{\Sigma}_{\mathbf{y}\mathbf{y}} &  \mathbf{\Sigma}_{\mathbf{y}\mathbf{x}} \\ \mathbf{\Sigma}_{\mathbf{x}\mathbf{y}}  & \mathbf{\Sigma}_{\mathbf{x}\mathbf{x}} \end{bmatrix}
\end{align*}
with $\mathbf{\Sigma}_{\mathbf{y}\mathbf{x}}^\mathrm{T}=\mathbf{\Sigma}_{\mathbf{x}\mathbf{y}}$

The OI or simple kriging (with zero mean) of $\mathbf{x}$ is given by:
\begin{align*}
\mathbf{x}^\star = \mathbf{\Sigma}_{\mathbf{x}\mathbf{y}}  \mathbf{\Sigma}_{\mathbf{y}\mathbf{y}}^{-1} \mathbf{y}
\end{align*}
Its error covariance matrix is:
\begin{align*}
\mathrm{Cov}(\mathbf{x}^\star-\mathbf{x}) = \Sigma_{\mathbf{x}\mathbf{x}} - \Sigma_{\mathbf{x}\mathbf{y}} \Sigma_{\mathbf{y}\mathbf{y}}^{-1} \Sigma_{\mathbf{y}\mathbf{x}}
\end{align*} 

We show that this formula can be rewritten:
\begin{align*}
\mathbf{x}^\star = -\mathbf{Q}_{\mathbf{x}\mathbf{x}}^{-1} \mathbf{Q}_{\mathbf{x}\mathbf{y}} \mathbf{y}
\end{align*}
Indeed :
\begin{align*}
\mathbf{\Sigma} \mathbf{Q} =  \begin{bmatrix} \mathbf{\Sigma}_{\mathbf{y}\mathbf{y}} &  \mathbf{\Sigma}_{\mathbf{y}\mathbf{x}} \\ \mathbf{\Sigma}_{\mathbf{x}\mathbf{y}}  & \mathbf{\Sigma}_{\mathbf{x}\mathbf{x}} \end{bmatrix}  \begin{bmatrix} \mathbf{Q}_{\mathbf{y}\mathbf{y}} &  \mathbf{Q}_{\mathbf{y}\mathbf{x}} \\ \mathbf{Q}_{\mathbf{x}\mathbf{y}}  & \mathbf{Q}_{\mathbf{x}\mathbf{x}} \end{bmatrix} =  \begin{bmatrix} \mathbf{I}_{\mathbf{y}\mathbf{y}} &  \mathbf{0}_{\mathbf{y}\mathbf{x}} \\ \mathbf{0}_{\mathbf{x}\mathbf{y}}  & \mathbf{I}_{\mathbf{x}\mathbf{x}} \end{bmatrix}
\end{align*}
Then,
\begin{align*}
&\mathbf{\Sigma}_{\mathbf{y}\mathbf{y}}\mathbf{Q}_{\mathbf{y}\mathbf{x}} + \mathbf{\Sigma}_{\mathbf{y}\mathbf{x}}\mathbf{Q}_{\mathbf{x}\mathbf{x}} &= \mathbf{0}_{\mathbf{y}\mathbf{x}}\\
&\mathbf{Q}_{\mathbf{y}\mathbf{x}} \mathbf{Q}_{\mathbf{x}\mathbf{x}}^{-1} + \mathbf{\Sigma}_{\mathbf{y}\mathbf{y}}^{-1}\mathbf{\Sigma}_{\mathbf{y}\mathbf{x}}&= \mathbf{0}_{\mathbf{y}\mathbf{x}}\\
\end{align*}
And by transposing:
\begin{align*}
& \mathbf{Q}_{\mathbf{x}\mathbf{x}}^{-1} \mathbf{Q}_{\mathbf{x}\mathbf{y}} + \mathbf{\Sigma}_{\mathbf{x}\mathbf{y}} \mathbf{\Sigma}_{\mathbf{y}\mathbf{y}}^{-1}= \mathbf{0}_{\mathbf{x}\mathbf{y}}\\
& \mathbf{\Sigma}_{\mathbf{x}\mathbf{y}}\mathbf{\Sigma}_{\mathbf{y}\mathbf{y}}^{-1}=-\mathbf{Q}_{\mathbf{x}\mathbf{x}}^{-1} \mathbf{Q}_{\mathbf{x}\mathbf{y}}
\end{align*}

Kriging $\mathbf{x}^\star$ of $\mathbf{y}$ is obtained by solving the linear system:
\begin{align*}
  \mathbf{x}^\star=-\mathbf{Q}_{\mathbf{x}\mathbf{x}}^{-1}\mathbf{Q}_{\mathbf{x}\mathbf{y}}\mathbf{y}
\end{align*}
and the error covariance matrix is:
\begin{align*}
\mathrm{Cov}(\mathbf{x}^\star-\mathbf{x}) = \mathbf{Q}_{\mathbf{x}\mathbf{x}}^{-1}
\end{align*}

This system is solved using Cholesky decomposition:

* a)  We first compute $\mathbf{L}$, the lower triangular Cholesky decomposition of $\mathbf{Q}_{\mathbf{x} \mathbf{x}}$:
\begin{align*}
  \mathbf{Q}_{\mathbf{x}\mathbf{x}}=\mathbf{L}\mathbf{L}^\mathrm{T}
\end{align*}
* b) We solve the linear system:
\begin{align*}
  \mathbf{L}\mathbf{X}=-\mathbf{Q}_{\mathbf{x}\mathbf{y}}\mathbf{y}
\end{align*}
* c) Then we solve the linear system:
\begin{align*}
  \mathbf{L}^\mathrm{T}\mathbf{x}^\star=\mathbf{X}
\end{align*}

These two linear systems can be solved by substitution because the matrix $\mathbf{L}$ is lower triangular.
Regarding the complexity of the linear systems involved in kriging/OI, let make the following remarks:

* In traditional geostatistics, we solve a system involving $\Sigma_{\mathbf{y}\mathbf{y}}$ which has the size of the dataset
* In the SPDE approach, we solve a system involving $Q_{\mathbf{x}\mathbf{x}}$ which has the size of state space. However, $\Sigma_{\mathbf{y}\mathbf{y}}$ is dense while $Q_{\mathbf{x}\mathbf{x}}$ is sparse. In the latter, we can use algorithms adapted to sparse matrices of complexity $\mathcal{O}(n^{3/2})$, while the general Cholesky algorithm is of complexity $\mathcal{O}(n^{3})$.

```r
# Generate a dataset
indexes_dat=NULL
for (j in 1:Ny){
    indexes_dat=c(indexes_dat,((Nx*(j-1))+10+j):(((Nx*(j-1))+10+j)+4))
    indexes_dat=c(indexes_dat,((Nx*(j-1))+10-j):(((Nx*(j-1))+10-j)-4))
}
indexes_dat=unique(indexes_dat[which(indexes_dat<=(Nx*Ny) & indexes_dat>0)])
# Target index (T)
indexes_tar = setdiff(1:(Nx*Ny),indexes_dat)
# Observation vector (ZD)
dat=Z[indexes_dat]
# Display
datPts=rep(NA,Nx*Ny)
datPts[indexes_dat]=dat
spplot(matrix(datPts,Nx),zlim=c(-2,2),title="Data")
```

```r
# Function for Kriging/OI 
## Inputs:  
## VTT: Cholesky decomposition of QTT
## QTD: Submatrix QTD
## dat: observations
## indexes_dat : index for observations
## Results: 
## Interpolated values
krig_Chol<-function(VTT, QTD,dat,indexes_dat){
  ## Second member
  RM=-QTD%*%dat
  #### Permutation
  RM=expand(VTT)$P%*%RM
  ## Solve first Linear system
  zT_k =solve(VTT,RM,system="L")
  ## Solve second Linear system
  zT_k =solve(VTT,zT_k,system="Lt")
  ## Inverse Permutation
  zT_k = t(expand(VTT)$P)%*%zT_k
    
  nbNodes=dim(QTD)[1]+dim(QTD)[2]
  z_k=rep(NA,nbNodes)
  z_k[indexes_dat]=dat
  z_k[-indexes_dat]=zT_k
  return(z_k)
}

# Precision matrix Q
Q=(dx*dy)*t(B)%*%B
# Submatrix QTD
QTD=Q[indexes_tar,indexes_dat]
# Cholesky factorization of QTT
VTT = Cholesky(Q[indexes_tar,indexes_tar],perm=TRUE,LDL=FALSE)
# Kriging
z_k=krig_Chol(VTT,QTD,dat,indexes_dat)
# Display
spplot(matrix(Z,Nx),title="Ground Truth",zlim=c(-2,2))
spplot(matrix(z_k,Nx),zlim=c(-2,2),title="Kriging")
contour(1:Nx,1:Ny,is.na(matrix(datPts,nrow=Nx)),nlevels=1,add=T,lwd=2,col='red',drawlabels=FALSE)
```

## II.4 Conditional simulation (by kriging)

Because OI acts by construction as a smoother of the hidden truth, an option to retrieve the original variability of the gaussian process is to run simulations conditionned by kriging. A simple though elegant solution in traditional geostatistics is to propose a conditional simulation $\mathbf{x}_x$ to be equal to:
\begin{align*}
 \mathbf{x}_c (\mathbf{s}) = \mathbf{x}^\star(\mathbf{s}) + (\mathbf{x}_{nc}(\mathbf{s}) - \mathbf{x}^\star_{nc}(\mathbf{s}))
\end{align*}

```r
# Non conditional simulation #2
znc2=simuNC(B,dx,dy,tau)
# Non conditional simulation #3
znc3=simuNC(B,dx,dy,tau)
# Kriging of simulation #2
znc_k2=krig_Chol(VTT,QTD,znc2[indexes_dat],indexes_dat)
# Kriging of simulation #2
znc_k3=krig_Chol(VTT,QTD,znc3[indexes_dat],indexes_dat)
# Conditional simulation 
zc2=z_k+(znc2-znc_k2)
zc3=z_k+(znc3-znc_k3)
# Display
spplot(matrix(datPts,Nx),title="Data",zlim=c(-2,2))
spplot(matrix(Z,Nx),title="Ground Truth",zlim=c(-2,2))
spplot(matrix(zc2,Nx),title="Conditional simulation #1",zlim=c(-2,2))
contour(1:Nx,1:Ny,is.na(matrix(datPts,nrow=Nx)),nlevels=1,add=T,lwd=2,col='red',drawlabels=FALSE)
spplot(matrix(zc3,Nx),title="Conditional simulation #2",zlim=c(-2,2))
contour(1:Nx,1:Ny,is.na(matrix(datPts,nrow=Nx)),nlevels=1,add=T,lwd=2,col='red',drawlabels=FALSE)
```

# III. Spatio-temporal extension

Let now consider the spatio-temporal extension to perform non-conditional simulations of a
random field $\mathbf{x}$ solution of the spatio-temporal SPDE:
\begin{eqnarray}
\label{isotropic_SPDE_st}
\frac{\partial{\mathbf{X}}}{\partial{t}}+(\kappa^2(\mathbf{s},t)-\Delta)^{\alpha/2} \mathbf{X}(\mathbf{s},t)=\tau \mathbf{Z}(\mathbf{s},t)
\end{eqnarray}

To ensure more regularity it is also possible to consider the noise process $\mathbf{Z}$ be white in time but with spatial dependence, such as a solution to an isotropic spatial SPDE for instance:
\begin{align*}
& \mathbf{Z}(\mathbf{s},t) = \mathbf{Z}_T(t) \otimes  \mathbf{W}_S(\mathbf{s})
\end{align*}
with
\begin{align}
\label{isotropic_spatial_SPDE}
& (\kappa_S^2 - \Delta)^{\alpha_S/2} \mathbf{W}_S(\mathbf{s}) = \mathbf{Z}_S(\mathbf{s})
\end{align}

This lead to the final model:
\begin{eqnarray}
\label{isotropic_SPDE_st_2}
\frac{\partial{\mathbf{X}}}{\partial{t}}+(\kappa^2(\mathbf{s},t)-\Delta)^{\alpha/2} \mathbf{X}(\mathbf{s},t)=\tau \mathbf{Z}_T(t) \otimes  \mathbf{W}_S(\mathbf{s})
\end{eqnarray}

We use a numerical implicit Euler scheme to discretize the space-time SPDE:
\begin{eqnarray*}
\frac{\mathbf{x}_{t+dt}-\mathbf{x}_{t}}{dt}+\mathbf{B}\mathbf{x}_{t+dt}=\frac{\tau}{\sqrt{dt}} \mathbf{w}_{t+dt}
\end{eqnarray*} 

where

* $\mathbf{x}_{t}$ the state space at time $t$
* $\mathbf{B}$ is still the discrete differential operator $(\kappa^2-\Delta)^{\alpha/2}$ 
* The random vectors $\lbrace \mathbf{w}_t$, $t>0 \rbrace$ are independent realizations of $\mathbf{W}_S$ the solution of a spatial isotropic SPDE

Let now develop the relation between $\mathbf{x}_{t+dt}$ using $\mathbf{x}_{t}$ and $\mathbf{z}_{t+dt}$ :
\begin{align}
\label{implicit_scheme}
& \frac{\mathbf{x}_{t+dt}-\mathbf{x}_{t}}{dt}+\mathbf{B}_{t+dt}\mathbf{x}_{t+dt}&=&\frac{\tau}{\sqrt{dt}} \mathbf{w}_{t+dt} \nonumber \\
& \mathbf{x}_{t+dt}-\mathbf{x}_{t} + dt\mathbf{B}_{t+dt}\mathbf{x}_{t+dt}&=&\tau\sqrt{dt} \mathbf{w}_{t+dt} \nonumber  \\
& (\mathbf{I}+dt\mathbf{B}_{t+dt})\mathbf{x}_{t+dt}&=&\mathbf{x}_{t}+\tau\sqrt{dt}\mathbf{w}_{t+dt} \nonumber  \\
& \mathbf{x}_{t+dt} &=&(\mathbf{I}+dt\mathbf{B}_{t+dt})^{-1}(\mathbf{x}_{t}+\tau\sqrt{dt}\mathbf{w}_{t+dt}) \nonumber  \\
& \mathbf{x}_{t+dt} &=& \mathbf{M}_{t+dt}(\mathbf{x}_{t}+\tau \sqrt{dt} \mathbf{w}_{t+dt}) &
\end{align}

From now on, let denote $\mathbf{M}_{t+dt}=(\mathbf{I}+dt\mathbf{B}_{t+dt})^{-1}$ the matrix operator to emulate the dynamical evolution of state $\mathbf{x}$ from time $t$ to $t+dt$. It will ease the link with formal data assimilation state space formulation.

## III.1 Isotropy

### III.1.a DF Discretized Laplacian

The discretized version of the Laplacian operator is the same that in the spatial case

```r
# Discretized Laplacian (differential operator)
## Inputs :  
## Nx : number of points in direction x
## Ny : number of points in direction y
## dx : lag in direction x
## dy : lag in direction y
## Results :
## Sparse matrix with size (Nx*Ny)x(Nx*Ny) 

DiffOperator<-function(Nx,Ny,dx,dy, kappa){
  
  nbNodes=Nx*Ny
  indices=1:nbNodes
  

  # We start by creating an array containing the indices of the non-zero elements
  # of A with associated values and we will use the R function sparseMatrix 
  # to create A from this array

  # Right neighbour ( !! remove last column)
  indicesVoisins=indices[-which(indices%%Nx==0)]
  # Matrix values
  k1=cbind(indicesVoisins,indicesVoisins+1, -1/dx^2)
  
  # Left neighbour ( !! remove first column)
  indicesVoisins=indices[-which(indices%%Nx==1)]
  # Matrix values
  k2=cbind(indicesVoisins,indicesVoisins-1, -1/dx^2)
  
  # Top neighbour  ( !! remove last line )
  indicesVoisins=indices[-which(indices>(Ny-1)*Nx)]
  # Matrix values
  k3=cbind(indicesVoisins,indicesVoisins+Nx, -1/dy^2)
  
  # Bottom neighbour ( !! remove first line )
  indicesVoisins=indices[-which(indices<Nx+1)]
  # Matrix values
  k4=cbind(indicesVoisins,indicesVoisins-Nx, -1/dy^2)
  
  # Central point
  k5=cbind(indices,indices,kappa^2+2*(1/dx^2+1/dy^2))

  # Stack all neighbours
  k=rbind(k1,k2,k3,k4,k5)
  
  # Create sparseMatrix
  return(sparseMatrix(i=k[,1],j=k[,2],x=k[,3],dims = c(nbNodes,nbNodes)))
  
}

# Test function
A=DiffOperator(10,10,1,1,0)
image(A)
```

### III.1.b Run the simulation

```r
# Discrete differential operator
A=DiffOperator(Nx,Ny,dx,dy,kappa)
B=pow_diff_operator(A,pow)

# Total time of the simulation
tps=100
# Time step
dt=1
# Number of steps
nbIter=tps/dt

# Cholesky decomposition of Id+A*A
AspChol=Cholesky(Diagonal(nbNodes)+dt*B,perm=TRUE,LDL=FALSE)

# Initialisation with random values
Z=rnorm(nbNodes)
if(!file.exists("./spplots")) dir.create("./spplots")
png("./spplots/frame%03d.png")
par(ask = FALSE)
for(i in 1:nbIter){ 
  ## simulate coloured noise
  noise = simuNC(A_w,dx,dy,tau_w)
  #noise = tau*rnorm(nbNodes)    
  ## Second member
  RM=(tau*1.4*sqrt(dt)/sqrt(dx*dy))*noise+Z
  ## Permutation
  RM=(expand(AspChol)$P)%*%RM
  ## Solve first Linear system
  Z=solve(AspChol, RM,system="L")
  ## Solve second Linear system
  Z=solve(AspChol, Z,system="Lt")
  ## Inverse Permutation
  Z=t(expand(AspChol)$P)%*%Z
  ## Display
  spplot(Z,zlim=c(-2,2),title=paste0("Time: t =",i*dt),display_vario=FALSE)
}
dev.off()
png_files <- sprintf("./spplots/frame%03d.png", 1:nbIter)
gif_file <- tempfile(tmpdir='.',fileext = ".gif")
gifski(png_files, gif_file,delay=.1)

print(gif_file)
display_html(paste0('<iframe src="',gif_file,'"  width=500, height=500></iframe>')) 
```

## III.2 Global anisotropy

The Laplacian operator $\Delta$ is invariant to translations and
rotations. As a consequence, we will get an isotropic space-time field. Then, let introduce some diffusion process through a local diffusion operator $\mathbf{H}(\mathbf{s},t)$ in the SPDE. In the case of global anisotropy, $\mathbf{H}$ is constant over space and time:

\begin{eqnarray*}
(\kappa^2-\nabla \cdot \mathbf{H}\nabla)^{\frac{\alpha}{2}}\mathbf{X}(\mathbf{s},t)=\tau\mathbf{Z}(\mathbf{s},t)
\end{eqnarray*}

### III.2.a DF Discretized Laplacian

```r
DiffOperator2<-function(Nx,Ny,dx,dy,H,kappa){
  
  H11=H[1,1] ; H12=H[1,2] ; H21=H[2,1] ; H22=H[2,2]  
  nbNodes=Nx*Ny
  indices=1:nbNodes
  
  # Right neighbour ( !! remove last column)
  indicesVoisins=indices[-which(indices%%Nx==0)]
  # Matrix values
  k1=cbind(indicesVoisins,indicesVoisins+1, -H11/(dx^2))
    
  # Left neighbour ( !! remove first column)
  indicesVoisins=indices[-which(indices%%Nx==1)]
  # Matrix values
  k2=cbind(indicesVoisins,indicesVoisins-1, -H11/(dx^2))

  # Top neighbour  ( !! remove last line )
  indicesVoisins=indices[-which(indices>(Ny-1)*Nx)]
  # Matrix values
  k3=cbind(indicesVoisins,indicesVoisins+Nx, -H22/(dy^2))
  
  # Bottom neighbour ( !! remove first line )
  indicesVoisins=indices[-which(indices<(Nx+1))]
  # Matrix values
  k4=cbind(indicesVoisins,indicesVoisins-Nx, -H22/(dy^2))
        
  # Top-right neighbour ( !! remove last column & last line )
  indicesVoisins=indices[-which(indices%%Nx==0 | indices>(Ny-1)*Nx)]
  # Matrix values
  k5=cbind(indicesVoisins,indicesVoisins+Nx+1, (H12+H21)/(4*dx*dy))
  
  # Top-left neighbour ( !! remove first column & last line)
  indicesVoisins=indices[-which(indices%%Nx==1  | indices>(Ny-1)*Nx)]
  # Matrix values
  k6=cbind(indicesVoisins,indicesVoisins+Nx-1, -(H12+H21)/(4*dx*dy))

  # Bottom-right neighbour ( !! remove last column & first line )
  indicesVoisins=indices[-which(indices%%Nx==0 | indices<Nx+1)]
  # Matrix values
  k7=cbind(indicesVoisins,indicesVoisins-Nx+1, -(H12+H21)/(4*dx*dy))
  
  # Bottom-left neighbour ( !! remove first column & first line)
  indicesVoisins=indices[-which(indices%%Nx==1 | indices<Nx+1)]
  # Matrix values
  k8=cbind(indicesVoisins,indicesVoisins-Nx-1, (H12+H21)/(4*dx*dy))
    
  # Central point
  k9=cbind(indices,indices,kappa^2+2*(H11/(dx^2)+H22/(dy^2)))
    
  # Stack all neighbours
  k=rbind(k1,k2,k3,k4,k5,k6,k7,k8,k9)

  # Create sparseMatrix
  return(sparseMatrix(i=k[,1],j=k[,2],x=k[,3],dims = c(nbNodes,nbNodes)))
  
}

# Test function
A=DiffOperator2(10,10,1,1,H=matrix(c(1,0,0,1),nrow=2),0)
image(A)
```

### III.2.e Spatio-temporal kriging

To compute space-time OI along a data assimilation window $\lbrace {\mathbf{y}_{t-Ldt},\cdots,\mathbf{y}_t}\rbrace$, we need to know how the process $\mathbf{x}_t$ at current time $t$ covariates with its past and future state. Then we are looking for $\mathbf{\Sigma}_{t+kdt,t}$ (covariance between past time and current time) and $\mathbf{\Sigma}_{t,t+kdt}$ (covariance between future time and current time).

According to Eq.(\ref{implicit_scheme}):
\begin{align*}
&\mathbf{\Sigma}_{t+dt,t}\\
&=\mathrm{Cov}\left\lbrace \mathbf{x}_{t+dt},\mathbf{x}_{t} \right\rbrace\\
&=\mathrm{Cov}\left\lbrace M_{t+dt}(\mathbf{x}_{t}+\tau dt\mathbf{z}_{t+dt}), \mathbf{x}_{t} \right\rbrace \\
&=M_{t+dt}\mathrm{Cov}\left\lbrace \mathbf{x}_{t}+\tau dt\mathbf{z}_{t+dt}, \mathbf{x}_{t} \right\rbrace \\
&=M_{t+dt} \mathbf{\Sigma}_{t,t} 
\end{align*}

Then,
\begin{align*}
\mathbf{Q}_{t+dt,t} &= \mathbf{Q}_{t,t} \mathbf{M}_{t+dt}^{-1} \\
\mathbf{Q}_{t,t+dt} &= \mathbf{M}_{t+dt}^{-1} \mathbf{Q}_{t,t} 
\end{align*}

Looking for covariances between between any past time and current time, we have:
\begin{align*}
&\mathbf{\Sigma}_{t+2dt,t} \\
&=\mathrm{Cov}\left\lbrace \mathbf{x}_{t+2dt},\mathbf{x}_{t} \right\rbrace \\
&=\mathrm{Cov}\left\lbrace \mathbf{M}_{t+2dt}(\mathbf{x}_{t+dt}+\tau dt\mathbf{z}_{t+2dt}),\mathbf{x}_{t} \right\rbrace \\
&=\mathrm{Cov}\left\lbrace \mathbf{M}_{t+2dt}(\mathbf{M}_{t+dt}(\mathbf{x}_{t}+\tau dt\mathbf{z}_{t+dt}) +dt\mathbf{z}_{t+dt},\mathbf{x}_{t} \right\rbrace \\
&=\mathbf{M}_{t+2dt}\mathbf{M}_{t+dt} \mathbf{\Sigma}_{t,t} 
\end{align*}

Then,
\begin{align*}
\mathbf{Q}_{t+2dt,t} &=  \mathbf{Q}_{t,t} \mathbf{M}_{t+dt}^{-1} \mathbf{M}_{t+2dt}^{-1}  \\
\mathbf{Q}_{t,t+2dt} &=  \mathbf{M}_{t+2dt}^{-1} \mathbf{M}_{t+dt}^{-1}  \mathbf{Q}_{t,t} 
\end{align*}

Leading to the iterative recursion:
\begin{align*}
\mathbf{Q}_{t+kdt,t} &= \mathbf{Q}_{t,t} \prod_{l=1}^{k} \mathbf{M}_{t+kdt}^{-1}\\
\mathbf{Q}_{t,t+kdt} &= \prod_{l=k}^{1} \mathbf{M}_{t+kdt}^{-1} \mathbf{Q}_{t,t}
\end{align*}

Let note that if operator $\mathbf{A}$ does not depend on time (then neither $\mathbf{Q}$), the recursion simplifies to:
\begin{align*}
\mathbf{Q}_{t+kdt,t} &= \mathbf{Q} \mathbf{M}^{-k}\\
\mathbf{Q}_{t,t+kdt} &= \mathbf{M}^{-k} \mathbf{Q}
\end{align*}

Also, the final model writes:
\begin{align*}
\mathbf{\Sigma}_{t+dt} & = \mathbf{M}\mathbf{\Sigma}_t\mathbf{M}^{\mathrm{T}} + \tau^2 \mathbf{M}\mathbf{Q}(W)^{-1}\mathbf{M}^{\mathrm{T}}
\end{align*}

The global covariance matrix of state sequences $\lbrace \mathbf{x}_1,\cdots,\mathbf{x}_{Ldt} \rbrace$ writes:
\begin{align*}
\mathbf{\Sigma}_G =
\begin{bmatrix} 
\mathbf{\Sigma}_1 & \mathbf{\Sigma}_1 \mathbf{M}_2 & \mathbf{\Sigma}_1 \mathbf{M}_2 \mathbf{M}_3 & \cdots & \mathbf{\Sigma}_1 \prod_{l=1}^{L} \mathbf{M}_{t+ldt} \\
\mathbf{M}_2 \mathbf{\Sigma}_1 & \mathbf{\Sigma}_2 & \mathbf{\Sigma}_2 \mathbf{M}_3 & \cdots & \mathbf{\Sigma}_2 \prod_{l=2}^{L} \mathbf{M}_{t+ldt}  \\
\mathbf{M}_3 \mathbf{M}_2 \mathbf{\Sigma}_1 & \mathbf{M}_3 \mathbf{\Sigma}_2 & \mathbf{\Sigma}_3 &  \cdots & \vdots \\
\vdots & \vdots & \vdots  & \ddots & \vdots \\
\prod_{l=L-1}^{2} \mathbf{M}_{t+ldt} \mathbf{\Sigma}_{1} &  \vdots & \vdots  & \mathbf{\Sigma}_{L-1} &  \mathbf{\Sigma}_{L-1} \mathbf{M}_{t+Ldt} \\
\prod_{l=L}^{2} \mathbf{M}_{t+ldt} \mathbf{\Sigma}_{1} &  \vdots & \vdots  & \mathbf{M}_{t+Ldt} \mathbf{\Sigma}_{L-1} &  \mathbf{\Sigma}_{L} 
\end{bmatrix}
\end{align*}

and the precision matrix writes:
\begin{align*}
\mathbf{Q}_G = \frac{1}{dt}[\mathbf{I}_{L} \otimes \mathbf{Q}(W)]
\begin{bmatrix} 
\mathbf{M}^{-2} & -\mathbf{M}^{-1} & 0 & \cdots & 0 \\
-\mathbf{M}^{-1}  & (\mathbf{M}^{-2}+\mathbf{I}) & -\mathbf{M}^{-1}  & \cdots & \vdots   \\
0 & -\mathbf{M}^{-1}  & (\mathbf{M}^{-2}+\mathbf{I}) & \cdots & 0 \\
\cdots & \cdots & \cdots  & \cdots & -\mathbf{M}^{-1}\\
0  &  \cdots & 0  & -\mathbf{M}^{-1} &  \mathbf{M}^{-2}
\end{bmatrix}
\end{align*}

```r
itime=50

## Global covariance matrix
M    = Diagonal(nbNodes)+dt*B 
M2   = M%*%M
M2pI = M2+Diagonal(nbNodes)
M0   = Matrix(matrix(0,nrow=nbNodes,ncol=nbNodes),sparse=TRUE)
replicateM0 <-function(N){
  return(do.call(cbind, replicate(N, M0, simplify=FALSE)))
}

DAW=2
nbBlocks = 2*DAW+1

QR = ((dx*dy)/tau_w^2)*t(A_w)%*%(A_w)
QG = NULL
# first line
QG_ = cbind(M2%*%QR,-M%*%QR,replicateM0(nbBlocks-2))
QG  = rbind(QG,QG_)
# loop
for (i in 2:(nbBlocks-1)){
  QG_ = cbind(replicateM0(i-2),-M%*%QR,M2%*%QR,-M%*%QR,replicateM0(nbBlocks-3-(i-2)))
  QG  = rbind(QG,QG_)
}
# last line
QG_ = cbind(replicateM0(nbBlocks-2),-M%*%QR,M2%*%QR)
QG  = rbind(QG,QG_)
# force symmetry
QG  = 1/2*(t(QG)+QG)

dat=NULL
indexes_dat=NULL
cpt=0
for (k in seq(-DAW,DAW,1)) {
   # Index for Submatrix QTD of Qg
   indexes_dat=c(indexes_dat,idD[[itime+k]]+cpt*nbNodes)
   # space-time observations for kriging
   dat=c(dat,Z_data_all[itime+k,idD[[itime+k]]])
   cpt=cpt+1
}
# Index for Submatrix QTT of Qg
indexes_tar = setdiff(1:(nbNodes*nbBlocks),indexes_dat)
# Cholesky factorization of QTT
VTT = Cholesky(QG[indexes_tar,indexes_tar],perm=TRUE,LDL=TRUE)
# Submatrix QTD of Q
QTD=QG[indexes_tar,indexes_dat] 
# observation at t=itime
dat0=Z_data_all[itime,idD[[itime]]]
image(QTD)
```

```r
krig_Chol_ST<-function(VTT, QTD,dat,indexes_dat,LDL=TRUE){
  ## Second member
  RM=-QTD%*%dat
  ## Permutation
  RM=expand(VTT)$P%*%RM
  ## Solve first Linear system
  zT_k =solve(VTT,RM,system="L")
  if(LDL==TRUE) {zT_k =solve(VTT,zT_k,system="D")} 
  ## Solve second Linear system
  zT_k =solve(VTT,zT_k,system="Lt")
  ## Inverse Permutation
  zT_k = t(expand(VTT)$P)%*%zT_k
  nbNodes_ST=dim(QTD)[1]+dim(QTD)[2]
  z_k=rep(NA,nbNodes_ST)
  z_k[indexes_dat]=dat
  z_k[-indexes_dat]=zT_k
  return(z_k)
}


# Spatio-temporal kriging
z_k_ST=krig_Chol_ST(VTT,QTD,dat,indexes_dat)[(DAW*nbNodes):(((DAW+1)*nbNodes)-1)]
znc=Z_all[itime,]
spplot(matrix(Z_data_all[itime,],nrow=Nx),zlim=c(-2,2),title='Data')
spplot(matrix(znc,nrow=Nx),zlim=c(-2,2),title='Simulation')
spplot(matrix(z_k_S,nrow=Nx),zlim=c(-2,2),title='Spatial kriging')
contour(1:Nx,1:Ny,is.na(matrix(Z_data_all[itime,],nrow=Nx)),nlevels=1,add=T,lwd=2,col='red',drawlabels=FALSE)
spplot(matrix(z_k_ST,nrow=Nx),zlim=c(-2,2),title='Spatio-temporal kriging')
contour(1:Nx,1:Ny,is.na(matrix(Z_data_all[itime,],nrow=Nx)),nlevels=1,add=T,lwd=2,col='red',drawlabels=FALSE)
```

## III.3 Local anisotropy

Let now introduce some local anisotropies. The underlying idea in this practical session would be to generate some artifical eddies.

\begin{eqnarray*}
(\kappa^2-\nabla \cdot \mathbf{H}(\mathbf{s},t)\nabla)^{\frac{\alpha}{2}}\mathbf{X}(\mathbf{s},t)=\tau\mathbf{Z}(\mathbf{s},t)
\end{eqnarray*}

### III.1.a Anisotropy field

```r
# Generate a periodic field to create some diffusion parameters
fxy<-function(x,y){
 return( ((2*resolution)/pi)*( (3/4)*sin(2*pi*x/(5*resolution))) + (1/4)*sin(2*pi*y/(5*resolution) ))
}
x=rep(1:Nx,Ny)
y=rep(1:Ny,each=Nx)
res=matrix(mapply(fxy,x,y),nrow=Nx)
grad=gradient(res,1,1)
u_=grad$X
v_=grad$Y
rotate_vector<-function(x1,y1,angle){
  x1 = as.vector(x1)
  y1 = as.vector(y1)
  x2=cos(angle)*x1-sin(angle)*y1
  y2=sin(angle)*x1+cos(angle)*y1
  return(cbind(x2,y2))
}
u=matrix(rotate_vector(u_,v_,-pi/2)[,1],nrow=Nx)
v=matrix(rotate_vector(u_,v_,-pi/2)[,2],nrow=Nx)
image.plot(res,col=viridis(128))
plot(range(50),range(50),type="n",xlim=c(0,Nx),ylim=c(0,Ny))
step_arrow=2
palette=viridis(128)
col=palette[floor(as.vector(sqrt(u^2+v^2)/max(sqrt(u^2+v^2)))*128)]
quiver(matrix(x,nrow=Nx)[seq(1,Nx,step_arrow),seq(1,Nx,step_arrow)],
       matrix(y,nrow=Nx)[seq(1,Ny,step_arrow),seq(1,Ny,step_arrow)],
       u[seq(1,Nx,step_arrow),seq(1,Nx,step_arrow)],
       v[seq(1,Ny,step_arrow),seq(1,Ny,step_arrow)],
       col=col,scale=0.5,length=0.1,angle=1,xlab="",ylab="",main="Vector field")
```

```r
# store all the diffusion parameters
library(abind)
Hs<-function(u,v,gamma,beta){
  nNodes = nrow(u)*ncol(u)
  H=array(NA,dim=c(2,2,nNodes))
  cpt=1
  for(i in 1:nrow(u)){
    for(j in 1:ncol(u)){
      vs=matrix(c(u[i,j],v[i,j]),nrow=2)
      Hs=(gamma*diag(2)) + beta*(vs%*%t(vs))
      H[,,cpt]=Hs
      cpt=cpt+1
    }
  }
  return(H)
}
Hs=Hs(u,v,1,25)
```

\begin{eqnarray}
H(\mathbf{s})=\gamma \mathbf{I}_2 + \beta\left[\mathbf{v}(\mathbf{s}) \mathbf{v}(\mathbf{s})^\mathrm{T}\right]
\end{eqnarray}

### III.3.b DF Discretized Laplacian

```r
DiffOperator3<-function(Nx,Ny,dx,dy,H,kappa){
  
  nbNodes=Nx*Ny
  indices=1:nbNodes
    
  # Right neighbour ( !! remove last column)
  indicesVoisins=indices[-which(indices%%Nx==0)]
  # Matrix values
  k1=cbind(indicesVoisins,indicesVoisins+1, -1*(H[1,1,indicesVoisins])/(dx^2))
  
  # Left neighbour ( !! remove first column)
  indicesVoisins=indices[-which(indices%%Nx==1)]
  # Matrix values
  k2=cbind(indicesVoisins,indicesVoisins-1, -1*(H[1,1,indicesVoisins])/(dx^2))

  # Top neighbour  ( !! remove last line )
  indicesVoisins=indices[-which(indices>(Ny-1)*Nx)]
  # Matrix values
  k3=cbind(indicesVoisins,indicesVoisins+Nx, -1*(H[2,2,indicesVoisins])/(dy^2))
  
  # Bottom neighbour ( !! remove first line )
  indicesVoisins=indices[-which(indices<(Nx+1))]
  # Matrix values
  k4=cbind(indicesVoisins,indicesVoisins-Nx, -1*(H[2,2,indicesVoisins])/(dy^2))
        
  # Top-right neighbour ( !! remove last column & last line )
  indicesVoisins=indices[-which(indices%%Nx==0 | indices>(Ny-1)*Nx)]
  # Matrix values
  k5=cbind(indicesVoisins,indicesVoisins+Nx+1, H[1,2,indicesVoisins]/(2*dx*dy))
    
  # Top-left neighbour ( !! remove first column & last line)
  indicesVoisins=indices[-which(indices%%Nx==1  | indices>(Ny-1)*Nx)]
  # Matrix values
  k6=cbind(indicesVoisins,indicesVoisins+Nx-1, -1*H[1,2,indicesVoisins]/(2*dx*dy))
    
  # Bottom-right neighbour ( !! remove last column & first line )
  indicesVoisins=indices[-which(indices%%Nx==0 | indices<Nx+1)]
  # Matrix values
  k7=cbind(indicesVoisins,indicesVoisins-Nx+1, -1*H[1,2,indicesVoisins]/(2*dx*dy))
  
  # Bottom-left neighbour ( !! remove first column & first line)
  indicesVoisins=indices[-which(indices%%Nx==1 | indices<Nx+1)]
  # Matrix values
  k8=cbind(indicesVoisins,indicesVoisins-Nx-1, H[1,2,indicesVoisins]/(2*dx*dy))
    
  # Central point
  k9=cbind(indices,indices,kappa^2+2*(H[1,1,indices]/(dx^2)+H[2,2,indices]/(dy^2)))
    
  # Stack all neighbours
  k=rbind(k1,k2,k3,k4,k5,k6,k7,k8,k9)

  # Create sparseMatrix
  return(sparseMatrix(i=k[,1],j=k[,2],x=k[,3],dims = c(nbNodes,nbNodes)))
  
}

# Test function
A=DiffOperator3(10,10,1,1,Hs,0)
image(A)
```

### III.1.c Run the simulation

```r
# Discrete differential operator
A=DiffOperator3(Nx,Ny,dx,dy,Hs,kappa)
B=pow_diff_operator(A,pow)

# Total time of the simulation
tps=100
# Time step
dt=1
# Number of steps
nbIter=tps/dt

# Cholesky decomposition of Id+A*A
AspChol=Cholesky(Diagonal(nbNodes)+dt*B,perm=FALSE,LDL=FALSE)

# Initialisation with random values
Z=rnorm(nbNodes)
Z_all = matrix(NA,nrow=nbIter, ncol=nbNodes)
if(!file.exists("./spplots")) dir.create("./spplots")
png("./spplots/frame%03d.png")
par(ask = FALSE)
for(i in 1:nbIter){ 
  ## simulate coloured noise
  #noise = simuNC(A_w,dx,dy,tau_w)
  noise = rnorm(nbNodes,0,1)
  ## Second member
  RM=(tau*5*sqrt(dt)/sqrt(dx*dy))*noise+Z
  ## Permutation
  RM=(expand(AspChol)$P)%*%RM
  ## Solve first Linear system
  Z=solve(AspChol, RM,system="L")
  ## Solve second Linear system
  Z=solve(AspChol, Z,system="Lt")
  ## Inverse Permutation
  Z=t(expand(AspChol)$P)%*%Z
  ## Display
  Z_all[i,]=Z
  spplot(Z,zlim=c(-2,2),title=paste0("Time: t =",i*dt),display_vario=FALSE)
}
dev.off()

png_files <- sprintf("./spplots/frame%03d.png", 1:nbIter)
gif_file <- tempfile(tmpdir='.',fileext = ".gif")
gifski(png_files, gif_file, delay=.1)

print(gif_file)
display_html(paste0('<iframe src="',gif_file,'"  width=500, height=500></iframe>'))
```

## V.II. Strategy #2: A formal link with (En)KF

In the end, we want an analysis $\mathbf{x}^a_{k}$ given by:
\begin{align*}
 \mathbf{x}^a_{k} & = \mathbf{x}^f_{k} + \mathbf{u}_{k}
\end{align*}
where $\mathbf{u}_{k+1}$ denotes a correction term, which finds the optimal compromise between the observations and the model prior.
From a geostatistical perspective, the optimal way of estimating $\mathbf{u}_{k}=\mathbf{x}^a_{k}-\mathbf{x}^f_{k}$ is given by:
\begin{align*}
 \mathbf{u}^\star_{k}=\mathbf{K}_{\mathbf{x}\mathbf{y}} \left(\mathbf{y}_k - \mathbf{H}_k \mathbf{x}_k 
\right)
\end{align*}
with $\mathbf{K}_{\mathbf{x}\mathbf{y}}$ the matrix of optimal weights of size $m \times p_k$ and $\left(\mathbf{y}_k - \mathbf{H}_k (\mathbf{x}_k) \right)$ are the innovations.
The matrix $\mathbf{K}$ is obtained by minimization of the estimation variance $\mathrm{Var}\left[ {u}^\star_{k} - {u}_{k} \right] $ leading to the so-called BLUE/(simple) kriging equations \cite{asch_data_2016,chiles_2012}:
\begin{align*}
  \mathbf{K}_{\mathbf{x}\mathbf{y}} & = \mathbf{\Sigma}_{\mathbf{x}\mathbf{y}}\mathbf{\Sigma}_{\mathbf{y}\mathbf{y}}^{-1}
\end{align*}
where, regarding the covariance-based equation we have $\mathbf{\Sigma}_{\mathbf{y}\mathbf{y}}$ the covariance of the innovations and $\mathbf{\Sigma}_{\mathbf{x}\mathbf{y}}$ the empirical cross-covariance matrices between $\mathbf{x}^f$ and $\mathbf{H}(\mathbf{x}^f)$ that writes:
\begin{align*}
 &\mathbf{\Sigma}_{\mathbf{y}\mathbf{y}} = E \left \lbrace \mathbf{d} \mathbf{d}^{\mathrm{T}} \right \rbrace &=& \mathbf{H}_k \mathbf{P}_k^f \mathbf{H}_k^{\mathrm{T}} + \mathbf{R} \\
 &\mathbf{\Sigma}_{\mathbf{x}\mathbf{y}} = E \left \lbrace \mathbf{x}^f \left(\mathbf{H}\mathbf{x}^f\right)^{\mathrm{T}} \right \rbrace &=& \mathbf{P}_k^f \mathbf{H}_k^{\mathrm{T}},  
\end{align*}

Going back to the global precision matrix, it writes:
\begin{align}
\label{covariance_OI_KF}
\mathbf{\Sigma} = \begin{bmatrix} \mathbf{\Sigma}_{\mathbf{y}\mathbf{y}} &  \mathbf{\Sigma}_{\mathbf{y}\mathbf{x}} \\ \mathbf{\Sigma}_{\mathbf{x}\mathbf{y}}  & \mathbf{\Sigma}_{\mathbf{x}\mathbf{x}} \end{bmatrix} =  =\begin{bmatrix} \mathbf{H}_k \mathbf{P}_k^f \mathbf{H}_k^{\mathrm{T}} + \mathbf{R} & \mathbf{H}_k \mathbf{P}_k^f \\  \mathbf{P}_k^f \mathbf{H}_k^{\mathrm{T}} & \mathbf{P}_k^f \end{bmatrix}
\end{align}

\begin{align*}
\mathbf{Q} =\begin{bmatrix} \mathbf{Q}_{\mathbf{y}\mathbf{y}} &  \mathbf{Q}_{\mathbf{y}\mathbf{x}} \\ \mathbf{Q}_{\mathbf{x}\mathbf{y}}  & \mathbf{Q}_{\mathbf{x}\mathbf{x}} \end{bmatrix} = \begin{bmatrix} \mathbf{R}_k^{-1} & -\mathbf{R}_k^{-1}\mathbf{H}_k \\ -\mathbf{H}_k^{\mathrm{T}}\mathbf{R}_k^{-1} & \mathbf{Q}^f_k+ \mathbf{H}_k^{\mathrm{T}}\mathbf{R}_k^{-1}\mathbf{H}_k \end{bmatrix} 
\end{align*}
where $\mathbf{Q}^f_k$ is nothing else than the precision matrix of state $\mathbf{x}_k$ at time $t_k$. This leads to the precision-based formulation of the Kalman Filter: 
\begin{align}
 \mathbf{x}^a_{k} & = \mathbf{x}^f_{k} -\mathbf{Q}_{\mathbf{x}\mathbf{x}}^{-1} \mathbf{Q}_{\mathbf{x}\mathbf{y}} \left( \mathbf{y}_k - \mathbf{H}_k (\mathbf{x}_k) \right) \nonumber \\
                  & = \mathbf{x}^f_{k} + \left( \mathbf{Q}^f_k+ \mathbf{H}^{\mathrm{T}}\mathbf{R}^{-1}\mathbf{H}  \right)^{-1} \mathbf{H}^{\mathrm{T}}\mathbf{R}^{-1} \left( \mathbf{y}_k - \mathbf{H}_k \mathbf{x}_k \right)
\end{align}
In a classic DA formulation, this type of formulation is not interesting because $\left( \mathbf{Q}^f_k+ \mathbf{H}_k^{\mathrm{T}}\mathbf{R}_k^{-1}\mathbf{H}_k  \right)^{-1}$ lies in state space and is not sparse. But within an SPDE-based formulation of $\mathbf{Q}^f_k$, Cholesky algorithm can again be involved to solve for the analysis step. In addition, it directly relates the precision matrix involved in the spatio-temporal simulation with the precision-based formulation of the Kalman gain.\\
